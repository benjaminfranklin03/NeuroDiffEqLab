```python
%matplotlib inline
import numpy as np
import jax
import jax.numpy as jnp
from jax import jit, grad, vmap, random, value_and_grad
from jax.experimental import ode
import optax
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
import imageio.v2 as imageio
import os

# setting style for seaborn plots
sns.set_theme(style="whitegrid")
```

```python
# ----------------------------
# Experiment Hyperparameters 
# ----------------------------
@dataclass
class ExperimentArgs:
    data_size: int = 1000       # total number of data points
    batch_size: int = 20        
    batch_time: int = 10        # number of time steps in each batch
    n_iters: int = 2000         # number of training iterations (i.e. epochs)
    test_freq: int = 20         # frequency of testing the model
    learning_rate: float = 0.001    # learning rate
    viz: bool = True            # Whether to visualize results
    mu: float = 1.0             # Parameter for the Van der Pol oscillator
    output_dir: str = 'training_plots'  # directory to save plots

args = ExperimentArgs()
```

```python
# ----------------------------
# Creating Synthetic Training Data
# ----------------------------

def vdp_oscillator(y, t, mu):
    """
    Van der Pol oscillator ODE system

    y: State vector 
    mu: parameter controlling nonlinearity
    --> returns derivatives
    """
    x, dx = y
    dy = mu * (1 - x**2) * dx - x
    return jnp.array([dx, dy])

def create_training_data(args):
    """
    generating training data by solving the ODE

    args (ExperimentArgs): experiment hyperparameters
    --> returns ts (time points), true_y0 (initial state), true_y (true state trajectories over time)
    """
    ts = jnp.linspace(0., 20., num=args.data_size)
    true_y0 = jnp.array([2., 0.])
    true_y = ode.odeint(lambda y, t: vdp_oscillator(y, t, args.mu), true_y0, ts)
    return ts, true_y0, true_y

```

```python
# ----------------------------
# Setting up the Neural ODE
# ----------------------------

def init_network_params(layer_sizes, key):
    """
    initializing neural network parameters
    """
    keys = random.split(key, len(layer_sizes) - 1)
    params = []
    for m, n, k in zip(layer_sizes[:-1], layer_sizes[1:], keys):
        # He initialization for weights
        W = random.normal(k, (n, m)) * jnp.sqrt(2 / m)
        b = jnp.zeros(n)
        params.append((W, b))
    return params

def neural_ode_func(params, y):
    """
    neural network that learns the vector field of the system

    params: Neural network parameters.
    y: Current state.
    
    --> returns output: Estimated derivatives
    """
    activations = y
    for W, b in params[:-1]:
        outputs = jnp.dot(W, activations) + b
        activations = jnp.tanh(outputs)
    W, b = params[-1]
    outputs = jnp.dot(W, activations) + b
    return outputs

def predict(params, y0, t):
    """
    predicting trajectory using the neural ODE model

    params: neural network parameters
    y0: initial state
    t: time points
    
    --> returns pred_y: Predicted trajectory over time
    """
    def ode_func(y, t):
        return neural_ode_func(params, y)
    pred_y = ode.odeint(ode_func, y0, t)
    return pred_y

def predict_batch(params, y0_batch, t):
    """
    Batch prediction for training

    params
    y0_batch: Batch of initial states.
    t: Time points.
        
    --> returns pred_y: predicted trajectories for the batch
    """
    def ode_func(y, t):
        return neural_ode_func(params, y)
    # vectorizing over the batch dimension
    ode_func_batched = vmap(lambda y0: ode.odeint(ode_func, y0, t), in_axes=(0,))
    # ODE integration for each initial state in the batch
    pred_y = ode_func_batched(y0_batch)  
    return jnp.swapaxes(pred_y, 0, 1)  #Swap axes to get shape [batch_time, batch_size, state_dim]
```

```python
# MSE Loss function
def mse_loss(pred_y, true_y):
    return jnp.mean((pred_y - true_y) ** 2)
```

```python
# ----------------------------
# Setting up all the plots for visualizations (01)
# ----------------------------

def get_batch(ts, true_y, args):
    """
    sampling a batch of training data from the generated trajectories
    
    ts: time points
    true_y: true trajectories
    args: ExperimentArgs
        
    --> returns batch_y0 (initial states for the batch), batch_t (time points for the batch), batch_y (true trajectories for the batch)
    """
    s = np.random.choice(np.arange(args.data_size - args.batch_time), args.batch_size, replace=False)
    batch_y0 = true_y[s]  
    batch_t = ts[:args.batch_time]  
    # stacking the trajectories for each time step in the batch
    batch_y = jnp.stack([true_y[s + i] for i in range(args.batch_time)], axis=0)  
    return batch_y0, batch_t, batch_y

def plot_loss(train_losses, test_losses, args):
    """
    plotting and saving the training and testing loss over iterations
    
    train_losses: training losses per iteration
    test_losses: testing losses at intervals
    args: ExperimentArgs
    """
    plt.figure(figsize=(12, 6))

    # we have to convert JAX arrays to NumPy arrays so that we can plot it
    train_losses_np = np.array([loss.item() for loss in train_losses])
    test_losses_np = np.array([loss.item() for loss in test_losses])

    sns.lineplot(x=range(len(train_losses_np)), y=train_losses_np, label='Training Loss', color='blue')
    sns.lineplot(x=np.arange(0, len(test_losses_np) * args.test_freq, args.test_freq), y=test_losses_np, label='Test Loss', color='orange')

    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    plt.title('Training and Test Loss Over Time')
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(args.output_dir, 'loss_plot.png'))
    plt.show()


def save_trajectory_plot(ts, true_y, pred_y, epoch, args):
    """
    Plots and saves the trajectories and phase portraits at a given epoch.
    
    ts: time points
    true_y: true trajectories
    pred_y: predicted trajectories
    epoch
    args: ExperimentArgs
    """
    plt.figure(figsize=(14, 5))
    ts_np = np.array(ts)
    true_y_np = np.array(true_y)
    pred_y_np = np.array(pred_y)

    # ploting trajectories
    plt.subplot(1, 2, 1)
    sns.lineplot(x=ts_np, y=true_y_np[:, 0], label='True x', color='green')
    sns.lineplot(x=ts_np, y=true_y_np[:, 1], label='True dx', color='blue')
    sns.lineplot(x=ts_np, y=pred_y_np[:, 0], label='Pred x', linestyle='--', color='green')
    sns.lineplot(x=ts_np, y=pred_y_np[:, 1], label='Pred dx', linestyle='--', color='blue')
    plt.xlabel('Time')
    plt.ylabel('States')
    plt.title(f'Trajectories (Epoch {epoch})')

    # plotting phase portrait
    plt.subplot(1, 2, 2)
    sns.lineplot(x=true_y_np[:, 0], y=true_y_np[:, 1], label='True Trajectory', color='green')
    sns.lineplot(x=pred_y_np[:, 0], y=pred_y_np[:, 1], label='Predicted Trajectory', linestyle='--', color='blue')
    plt.xlabel('x')
    plt.ylabel('dx/dt')
    plt.title(f'Phase Portrait (Epoch {epoch})')

    plt.tight_layout()
    filename = os.path.join(args.output_dir, f'epoch_{epoch:04d}.png')
    plt.savefig(filename)
    plt.close()

# generating a GIF from the saved trajectory plots
def create_gif(args):
    images = []
    filenames = sorted([f for f in os.listdir(args.output_dir) if f.startswith('epoch_') and f.endswith('.png')])
    for filename in filenames:
        images.append(imageio.imread(os.path.join(args.output_dir, filename)))
    gif_path = os.path.join(args.output_dir, 'training_evolution.gif')
    imageio.mimsave(gif_path, images, fps=2)
    print(f"GIF saved to {gif_path}")

```

```python
# ----------------------------
# Setting up all the plots for visualizations (02)
# ----------------------------

def plot_vector_fields(params, args):
    """
    plotting the true and learned vector fields of the system
    
    params: Neural network parameters.
    args: ExperimentArgs
    """
    # creating a grid of points over the phase space
    x = np.linspace(-3, 3, 20)
    y = np.linspace(-3, 3, 20)
    X, Y = np.meshgrid(x, y)
    XY = np.stack([X.flatten(), Y.flatten()], axis=0).T  

    # defining functions to compute the vector fields
    def true_vector_field(y):
        return vdp_oscillator(y, 0, args.mu)

    def learned_vector_field(y):
        return neural_ode_func(params, y)

    # computing vector fields at each point in the grid
    true_dy = vmap(true_vector_field)(XY)
    learned_dy = vmap(learned_vector_field)(XY)

    # reshaping the results for plotting
    U_true = true_dy[:, 0].reshape(X.shape)
    V_true = true_dy[:, 1].reshape(Y.shape)
    U_learned = learned_dy[:, 0].reshape(X.shape)
    V_learned = learned_dy[:, 1].reshape(Y.shape)

    # plotting the true and learned vector fields side by side
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # true vector field
    axes[0].streamplot(X, Y, U_true, V_true, color='blue')
    axes[0].set_title('True Vector Field')
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('dx/dt')
    axes[0].set_aspect('equal')

    # learned vector field
    axes[1].streamplot(X, Y, U_learned, V_learned, color='red')
    axes[1].set_title('Learned Vector Field')
    axes[1].set_xlabel('x')
    axes[1].set_ylabel('dx/dt')
    axes[1].set_aspect('equal')

    plt.tight_layout()
    plt.show()


def plot_multiple_trajectories(params, args):
    """
    Plots trajectories from different initial conditions to test generalization of the model
    
    params: neural network parameters
    args: ExperimentArgs
    """
    # defining a set of initial conditions
    initial_conditions = np.array([
        [2.0, 0.0],
        [1.0, 1.0],
        [-1.0, -1.0],
        [0.5, -0.5],
        [-2.0, 1.5]
    ])
    ts = np.linspace(0, 20, 200)

    plt.figure(figsize=(8, 6))
    for y0 in initial_conditions:
        # true trajectory
        true_y = ode.odeint(lambda y, t: vdp_oscillator(y, t, args.mu), y0, ts)
        true_y_np = np.array(true_y)

        # predicted trajectory
        pred_y = predict(params, y0, ts)
        pred_y_np = np.array(pred_y)

        # plotting the trajectories
        plt.plot(true_y_np[:, 0], true_y_np[:, 1], label='True Trajectory', color='blue', alpha=0.5)
        plt.plot(pred_y_np[:, 0], pred_y_np[:, 1], '--', label='Predicted Trajectory', color='red', alpha=0.5)

    plt.title('Trajectories from Different Initial Conditions')
    plt.xlabel('x')
    plt.ylabel('dx/dt')
    plt.legend(['True Trajectory', 'Predicted Trajectory'])
    plt.grid(True)
    plt.show()

def plot_error_heatmap(params, args):
    """
    plotting a heatmap of the error magnitude between the true and learned vector fields across the phase space
    
    params: neural network parameters
    args: ExperimentArgs
    """
    x = np.linspace(-3, 3, 50)
    y = np.linspace(-3, 3, 50)
    X, Y = np.meshgrid(x, y)
    XY = np.stack([X.flatten(), Y.flatten()], axis=1)

    # computing the true and learned vector fields
    def true_vector_field(y):
        return vdp_oscillator(y, 0, args.mu)

    def learned_vector_field(y):
        return neural_ode_func(params, y)

    true_dy = vmap(true_vector_field)(XY)
    learned_dy = vmap(learned_vector_field)(XY)

    # calculating the error magnitude
    error = np.linalg.norm(true_dy - learned_dy, axis=1)
    error_grid = error.reshape(X.shape)

    # plotting the error heatmap
    plt.figure(figsize=(8, 6))
    contour = plt.contourf(X, Y, error_grid, levels=50, cmap='viridis')
    plt.colorbar(contour, label='Error Magnitude')
    plt.title('Error Heatmap Across Phase Space')
    plt.xlabel('x')
    plt.ylabel('dx/dt')
    plt.axis('equal')
    plt.show()
```

```python
# ----------------------------
# Main function
# ----------------------------

def main():
    # preparing output directory
    if not os.path.exists(args.output_dir):
        os.makedirs(args.output_dir)

    ts, true_y0, true_y = create_training_data(args)

    # initializing neural network parameters and the optimizer
    layer_sizes = [2, 50, 50, 2]
    params = init_network_params(layer_sizes, random.PRNGKey(0))
    optimizer = optax.adam(args.learning_rate) #using the adam optimizer like in the Neural ODE 01 implementation
    opt_state = optimizer.init(params)
    
    # defining the training update function with JIT compilation for performance
    @jit #jit compile
    def update(params, opt_state, batch_y0, batch_t, batch_y):
        """
        Performs a single training step.

        here we calculate the loss, compute gradients, and update the
        neural network parameters using the optimizer. I also used the JIT compiler for
        performance, since this stepwill be called repeatedly during training
        """
        def loss_fn(params):
            pred_y = predict_batch(params, batch_y0, batch_t) #predicting batch trajectories
            return mse_loss(pred_y, batch_y) #computing MSE
        loss, grads = value_and_grad(loss_fn)(params) #computing loss and gradients
        #updating parameters using the optimizer
        updates, opt_state = optimizer.update(grads, opt_state, params)
        params = optax.apply_updates(params, updates)
        return params, opt_state, loss

    train_losses = []
    test_losses = []

    # training loop
    for i in range(args.n_iters):
        batch_y0, batch_t, batch_y = get_batch(ts, true_y, args)
        params, opt_state, loss = update(params, opt_state, batch_y0, batch_t, batch_y)
        train_losses.append(loss)

        # saving plot at testing intervals
        if i % args.test_freq == 0:
            # predicting over the entire time span
            pred_y = predict(params, true_y0, ts)
            test_loss = mse_loss(pred_y, true_y)
            test_losses.append(test_loss)
            print(f"Iteration {i} | Training Loss: {loss:.6f} | Test Loss: {test_loss:.6f}")

            if args.viz:
                save_trajectory_plot(ts, true_y, pred_y, i, args)

    # plot of final loss history
    plot_loss(train_losses, test_losses, args)

    # creating GIF from saved plots
    create_gif(args)

    # visualizations (from part 2)
    pred_y = predict(params, true_y0, ts)
    pred_y = np.array(pred_y)
    true_y_np = np.array(true_y)

    #  vector field plot
    plot_vector_fields(params, args)

    #  multiple trajectories plot
    plot_multiple_trajectories(params, args)
    
    #  error heatmap across phase space
    plot_error_heatmap(params, args)


if __name__ == "__main__":
    main()
```
